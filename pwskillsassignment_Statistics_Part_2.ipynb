# **Statistics Part 2**

1. What is hypothesis testing in statistics?

Hypothesis testing is a statistical method used to make decisions about a population based on sample data. It helps determine whether there is enough evidence to reject a null assumption about the population.

2. What is the null hypothesis, and how does it differ from the alternative hypothesis?

Null hypothesis (H₀): Assumes no effect, no difference, or status quo.

Alternative hypothesis (H₁ or Ha): Assumes there is an effect or difference.
Example: H₀: μ = 50, H₁: μ ≠ 50.

3. What is the significance level in hypothesis testing, and why is it important?

Significance level (α): Probability of rejecting H₀ when it is actually true (Type I error).

Common values: 0.05, 0.01.

It sets the threshold for statistical evidence.

4. What does a P-value represent in hypothesis testing?


The P-value is the probability of observing results at least as extreme as the sample results, assuming H₀ is true.

5. How do you interpret the P-value in hypothesis testing?

If P ≤ α → Reject H₀ (evidence supports H₁).

If P > α → Fail to reject H₀ (insufficient evidence).

6. What are Type 1 and Type 2 errors in hypothesis testing?

Type I error (α): Rejecting H₀ when it’s true.

Type II error (β): Failing to reject H₀ when it’s false.

7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?

One-tailed: Tests for effect in one direction (>, <).

Two-tailed: Tests for effect in both directions (≠).

8. What is the Z-test, and when is it used in hypothesis testing?

A Z-test checks if sample mean differs from population mean when population variance is known and sample size is large (n > 30).

9. How do you calculate the Z-score, and what does it represent in hypothesis testing?

Represents how many standard errors the sample mean is away from the population mean.

10. What is the T-distribution, and when should it be used instead of the normal distribution?

T-distribution is used when the population standard deviation is unknown and sample size is small (n < 30).

11. What is the difference between a Z-test and a T-test?

Z-test: Population variance known, large samples.

T-test: Population variance unknown, small samples.

12. What is the T-test, and how is it used in hypothesis testing?

T-test compares sample mean(s) with population mean or between two groups, accounting for variability in small samples.

13. What is the relationship between Z-test and T-test in hypothesis testing?

Both test mean differences. T-test approximates Z-test as sample size increases.

14. What is a confidence interval, and how is it used to interpret statistical results?

A confidence interval gives a range of values within which the true population parameter is likely to lie with a certain confidence (e.g., 95%).



15. What is the margin of error, and how does it affect the confidence interval?

Margin of error = half-width of confidence interval. Larger margin → wider interval, lower precision.

16. How is Bayes' Theorem used in statistics, and what is its significance?

Bayes’ theorem updates probability of a hypothesis based on new evidence.

17. What is the Chi-square distribution, and when is it used?

Distribution of sum of squared standard normal variables. Used in tests of independence and goodness-of-fit.

18. What is the Chi-square goodness of fit test, and how is it applied?

Checks if observed frequencies match expected frequencies.

Ratio of two chi-square distributions. Used in variance comparison and ANOVA.

20. What is an ANOVA test, and what are its assumptions?

ANOVA (Analysis of Variance) compares means across multiple groups.
Assumptions: independence, normality, equal variances.

21. What are the different types of ANOVA tests?

One-way ANOVA

Two-way ANOVA (with/without interaction)

Repeated measures ANOVA

22. What is the F-test, and how does it relate to hypothesis testing?

F-test compares variances (ratio of variances). In ANOVA, it checks if at least one group mean differs significantly.

3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.

```python
def bernoulli_pdf(p, x):
    """PDF of Bernoulli distribution"""
    if x not in [0, 1]:
        return 0
    return p**x * (1-p)**(1-x)

p = 0.5
print("P(X=1):", bernoulli_pdf(p, 1))
print("P(X=0):", bernoulli_pdf(p, 0))
```

1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and
interpret the results.

```python
import numpy as np
from scipy.stats import norm

# Sample data
sample = np.random.normal(52, 10, 30)
pop_mean = 50
sigma = 10

z = (np.mean(sample) - pop_mean) / (sigma / np.sqrt(len(sample)))
p_value = 2 * (1 - norm.cdf(abs(z)))

print("Z-statistic:", z)
print("P-value:", p_value)

if p_value < 0.05:
    print("Reject H0: Significant difference")
else:
    print("Fail to reject H0: No significant difference")
```

2. Simulate random data to perform hypothesis testing and calculate the corresponding P-value using Python.

3. Implement a one-sample Z-test using Python to compare the sample mean with the population mean.

```python
def one_sample_ztest(data, mu, sigma):
    z = (np.mean(data) - mu) / (sigma / np.sqrt(len(data)))
    p_value = 2 * (1 - norm.cdf(abs(z)))
    return z, p_value

data = np.random.normal(52, 10, 30)
print(one_sample_ztest(data, 50, 10))
```

14. Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution.

```python
from scipy.stats import chisquare

obs = [16, 18, 16, 14, 12, 12]
# Adjust expected frequencies so they sum to the same total as obs
exp = [15, 15, 15, 15, 15, 15]
exp = [e * (sum(obs) / sum(exp)) for e in exp]

chi2, p = chisquare(f_obs=obs, f_exp=exp)
print("Goodness of Fit Test Statistic:", chi2)
print("P-value:", p)
```

```python
p1, n1 = 0.55, 100
p2, n2 = 0.60, 120
p_pool = (p1*n1 + p2*n2) / (n1+n2)

z = (p1 - p2) / np.sqrt(p_pool * (1-p_pool) * (1/n1 + 1/n2))
print("Z-test statistic for proportions:", z)
```
